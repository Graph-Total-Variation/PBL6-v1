```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python

# c hai bc nh gc v sau khi denoise
image_original = cv2.imread('dataset/brain/ref/1.png', cv2.IMREAD_GRAYSCALE)
image_denoised = cv2.imread('dataset/brain/noisy/1.png', cv2.IMREAD_GRAYSCALE)

intensity_original = image_original.ravel()
intensity_denoised = image_denoised.ravel()
# To trc X (X Direction)
x_direction = list(range(len(intensity_original)))

# V biu  Intensity-Direction cho c hai nh
plt.figure(figsize=(12, 6))
plt.plot(x_direction, intensity_original, label='Original Image', linewidth=1, color='blue', linestyle='--')
plt.plot(x_direction, intensity_denoised, label='Denoised Image', linewidth=1, color='black', linestyle='--')
plt.title('Intensity-Direction')
plt.xlabel('X Direction')
plt.ylabel('Intensity')
plt.legend()
plt.grid(True)
plt.show()
```

```python
image_original = cv2.imread('dataset/dataset_structure/ref/1.png', cv2.IMREAD_GRAYSCALE)
image_noisy = cv2.imread('dataset/dataset_structure/noisy/1.png', cv2.IMREAD_GRAYSCALE)
image_denoised = cv2.imread('result/1.png', cv2.IMREAD_GRAYSCALE)

row = 200  # V d: chn dng th 100
intensity_original = image_original[row, :]
intensity_noisy = image_noisy[row, :]
intensity_denoised = image_denoised[row, :]
# To trc X (ch s ct)
x_direction = range(intensity_original.shape[0])

fig, axes = plt.subplots(2, 3, figsize=(12, 6))
axes[0,0].imshow(image_original,cmap='gray')
axes[0,0].set_title('grouth truth')
axes[0,0].axis('off')

axes[0,1].imshow(image_noisy,cmap='gray')
axes[0,1].set_title('noisy\nPSNR:{:.2f}\nSSIM:{:.2f}'.format(12,12))
axes[0,1].axis('off')

axes[0,2].imshow(image_denoised,cmap='gray')
axes[0,2].set_title('denoise\nPSNR:{:.2f}\nSSIM:{:.2f}'.format(11,11))
axes[0,2].axis('off')

axes[1,0].plot(x_direction, intensity_original, color='blue', label='Original Image', linestyle='--', linewidth=2)
axes[1,0].set_xlabel('X Direction')
axes[1,0].set_ylabel('Intensity')
axes[1,0].legend()
axes[1,0].grid(True)

axes[1,1].plot(x_direction, intensity_original, color='blue', label='Original Image', linestyle='--', linewidth=1)
axes[1,1].plot(x_direction, intensity_noisy, color='black', label='Noisy Image', linestyle='--', linewidth=2)
axes[1,1].set_xlabel('X Direction')
axes[1,1].set_ylabel('Intensity')
axes[1,1].legend()
axes[1,1].grid(True)

axes[1,2].plot(x_direction, intensity_original, color='blue', label='Original Image', linestyle='--', linewidth=1)
axes[1,2].plot(x_direction, intensity_denoised, color='black', label='Denoise Image', linestyle='--', linewidth=2)
axes[1,2].set_xlabel('X Direction')
axes[1,2].set_ylabel('Intensity')
axes[1,2].legend()
axes[1,2].grid(True)
plt.tight_layout() 
```
