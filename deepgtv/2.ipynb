```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Đọc hai bức ảnh gốc và sau khi denoise
image_original = cv2.imread('dataset/brain/ref/1.png', cv2.IMREAD_GRAYSCALE)
image_denoised = cv2.imread('dataset/brain/noisy/1.png', cv2.IMREAD_GRAYSCALE)

intensity_original = image_original.ravel()
intensity_denoised = image_denoised.ravel()
# Tạo trục X (X Direction)
x_direction = list(range(len(intensity_original)))

# Vẽ biểu đồ Intensity-Direction cho cả hai ảnh
plt.figure(figsize=(12, 6))
plt.plot(x_direction, intensity_original, label='Original Image', linewidth=1, color='blue', linestyle='--')
plt.plot(x_direction, intensity_denoised, label='Denoised Image', linewidth=1, color='black', linestyle='--')
plt.title('Intensity-Direction')
plt.xlabel('X Direction')
plt.ylabel('Intensity')
plt.legend()
plt.grid(True)
plt.show()
```

```python
image_original = cv2.imread('dataset/dataset_structure/ref/1.png', cv2.IMREAD_GRAYSCALE)
image_noisy = cv2.imread('dataset/dataset_structure/noisy/1.png', cv2.IMREAD_GRAYSCALE)
image_denoised = cv2.imread('result/1.png', cv2.IMREAD_GRAYSCALE)
# Chọn dòng hoặc cột bạn muốn plot
row = 200  # Ví dụ: chọn dòng thứ 100
intensity_original = image_original[row, :]
image_noisy = image_noisy[row, :]
intensity_denoised = image_denoised[row, :]
# Tạo trục X (chỉ số cột)
x_direction = range(intensity_original.shape[0])
# Tạo subplot
plt.figure(figsize=(12, 6))
plt.subplot(131)  # Dòng 1, cột 2, biểu đồ 1
plt.plot(x_direction, intensity_original, color='blue', label='Original Image', linestyle='--', linewidth=2)
plt.title('Intensity-Direction (Original Image)')
plt.xlabel('X Direction')
plt.ylabel('Intensity')
plt.legend()
plt.grid(True)

plt.subplot(132)  # Dòng 1, cột 2, biểu đồ 2
plt.plot(x_direction, intensity_original, color='blue', label='Original Image', linestyle='--', linewidth=1)
plt.plot(x_direction, image_noisy, color='black', label='Noisy Image', linestyle='--', linewidth=2)
plt.xlabel('X Direction')
plt.ylabel('Intensity')
plt.legend()
plt.grid(True)

plt.subplot(133)  # Dòng 1, cột 2, biểu đồ 2
plt.plot(x_direction, intensity_original, color='blue', label='Original Image', linestyle='--', linewidth=1)
plt.plot(x_direction, intensity_denoised, color='black', label='Denoised Image', linestyle='--', linewidth=2)
plt.xlabel('X Direction')
plt.ylabel('Intensity')
plt.legend()
plt.grid(True)


plt.tight_layout()  # Đảm bảo không bị trùng lấn giữa hai biểu đồ
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(x_direction, intensity_original, color='blue', label='Original Image', linestyle='--', linewidth=1)
plt.plot(x_direction, intensity_denoised, color='black', label='Denoised Image', linestyle='--', linewidth=2)
plt.xlabel('X Direction')
plt.ylabel('Intensity')
plt.legend()
plt.grid(True)

plt.tight_layout()  # Đảm bảo không bị trùng lấn giữa hai biểu đồ
plt.show()
```
